# **Brakujące wartości**

**Brakujące wartości** to rekordy, które są nieobecne w zestawie danych.

**Przyczyny występowania**:
* Błąd człowieka podczas wprowadzania danych​

* Ochrona prywatności (dane wrażliwe)
* Brak odpowiedzi w ankietach

Problemy techniczne przy zbieraniu danych

Jest to jeden z **najczęściej występujących problemów** w analizie danych i kluczowy etap wstępnego przetwarzania (data preprocessing).

````python
# Definicja DataFrame
df_student = pd.DataFrame({
    'imie': ['Anna', 'Piotr', 'Maria', 'Jan', 'Katarzyna', 'Tomasz', 'Magdalena', 'Paweł', 'Ewa', 'Michał'],
    'wiek': [20, 22, 21, np.nan, 23, 20, 22, np.nan, 21, 23],
    'ocena_matematyka': [4.5, np.nan, 5.0, 4.0, 3.5, 4.5, 5.0, 3.5, np.nan, 4.0],
    'ocena_fizyka': [5.0, 4.5, np.nan, 4.0, 4.5, np.nan, 5.0, 4.0, 4.5, np.nan],
    'ocena_informatyka': [4.0, 3.5, 4.5, np.nan, 5.0, 3.5, 4.5, 4.0, np.nan, 4.5],
    'miasto': ['Warszawa', 'Kraków', 'Wrocław', 'Poznań', 'Gdańsk', np.nan, 'Warszawa', 'Kraków', 'Wrocław', 'Poznań']
})
````

* **Usuwanie rekordów** zawierających brakujące wartości.
* **Ręczne uzupełnianie** brakujących wartości.

Uzupełnianie **najbardziej prawdopodobną wartością** przy użyciu modeli uczenia maszynowego, takich jak regresja, drzewa decyzyjne czy algorytm KNN.

## **Uzupełnienie średnią lub modą​**

W pandas *dataFrame* możemy uzupełnić brakujące wartości za pomocą funkcji `fillna()`.
* Przyjmuje ona jedną wartość, która będzie wstawiana w pustych pozycjach lub zamiast wartości NaN.

Uzupełnia wszystkie brakujące wartości w kolumnie age średnią obliczoną z tejże kolumny

```data['age'] = data.age.fillna(data.age.mean())​```

Uzupełnia wszystkie brakujące wartości w kolumnie income medianą obliczoną z tejże kolumny

```data['income']=data.income.fillna(data.income.median())```

Zastępuje wszystkie brakujące wartości w kolumnie gender (kolumna kategorii) dominantą wyliczoną z tejże kolumny​

```data['gender']=data['gender'].fillna(data['gender'].mode()[0])​```

## 1. Wczytanie danych i analiza braków

W pliku braki są zapisane na **wiele sposobów**, zależnie od kolumny:

| Rodzaj | Przykłady w danych |
|--------|--------------------|
| tekstowe znaczniki | `NA`, `N/A`, `brak`, `?` |
| sentinele liczbowe | `-999` (piętro, cena_m2), `-1` (cena_m2) |
| wartości nielogiczne | `powierzchnia = 0` (mieszkanie nie ma 0 m²) |
| błąd wpisu / obcy język | `stan = 很好` (chiń. „bardzo dobrze") |

Dlatego samo `read_csv` z domyślnym `na_values` **nie wychwyci wszystkiego** — trzeba
obsłużyć braki kolumna po kolumnie, bo mają różne znaczenie.

````python
# Wczytujemy plik, od razu zamieniając tekstowe znaczniki braku na NaN
braki_tekstowe = ['NA', 'N/A', 'brak', 'brak danych', '?', '-', '']
df = pd.read_csv('nieruchomosci.csv', na_values=braki_tekstowe)

print("Kształt danych:", df.shape)
print("\nTypy kolumn:")
print(df.dtypes)
df.head()
````

````python
# Braki wychwycone przez na_values (znaczniki tekstowe -> NaN):
print("Braki po wczytaniu (tylko znaczniki tekstowe):")
print(df.isna().sum())

# Ale w danych zostały jeszcze UKRYTE braki: sentinele liczbowe i wartości nielogiczne
print("\n--- ukryte braki / wartości podejrzane ---")
print("powierzchnia == 0 :", (df['powierzchnia'] == 0).sum(), "(mieszkanie nie ma 0 m²)")
print("pietro == -999    :", (df['pietro'] == -999).sum(), "(sentinel braku)")
print("cena_m2 <= 0      :", (df['cena_m2'] <= 0).sum(), "(-1 oraz -999)")
print("\nUnikalne wartości 'stan':", list(df['stan'].dropna().unique()))
print("  -> '很好' to błąd wpisu (chiń. 'bardzo dobrze'), a nie osobna kategoria")
````

## 2. Typy zmiennych i strategia dla każdej kolumny

**Typy zmiennych:**
- **numeryczne ciągłe:** `powierzchnia`, `cena_m2`, `cena_total`
- **numeryczne dyskretne:** `liczba_pokoi`, `rok_budowy`, `pietro`, `liczba_pieter`
- **kategorialne:** `miasto`, `dzielnica`, `stan`, `parking` (Tak/Nie), `balkon` (Tak/Nie)

**Nie** — braki nie powinny być obsłużone tak samo. Strategia zależy od typu i znaczenia kolumny:

| Kolumna | Problem | Rozwiązanie | Uzasadnienie |
|---------|---------|-------------|--------------|
| `stan` | `很好` = błąd wpisu | zamiana na `Bardzo dobry` | to nie brak, tylko literówka/obcy język |
| `powierzchnia` | `0`, `NA`, `brak` | **przelicz** z `cena_total / cena_m2`, reszta → mediana | zależność biznesowa: cena_total ≈ powierzchnia × cena_m2 |
| `cena_m2` | `-1`, `-999` | **przelicz** z `cena_total / powierzchnia` | jw. — dokładniejsze niż średnia |
| `rok_budowy` | `NA`, `brak` | mediana (int) | rok musi być liczbą całkowitą, rozkład skośny → mediana |
| `pietro` | `-999`, `brak` | mediana (int) | dyskretna; `0` (parter) jest POPRAWNE, zostaje |
| `parking`, `balkon` | `?` | dominanta | zmienne kategorialne binarne |

**Kluczowa obserwacja:** `powierzchnia × cena_m2 = cena_total` (sprawdź: 65 × 12500 = 812500 ✓).
Dzięki temu brakującą wartość można **odtworzyć dokładnie** z dwóch pozostałych, zamiast zgadywać średnią.

````python
# 3. Zamiana sentineli i wartości nielogicznych na NaN + naprawa błędu wpisu
df['powierzchnia'] = df['powierzchnia'].replace(0, np.nan)      # 0 m² niemożliwe
df['pietro']       = df['pietro'].replace(-999, np.nan)         # -999 = brak
df['cena_m2']      = df['cena_m2'].replace([-1, -999], np.nan)  # ujemna cena niemożliwa
df['stan']         = df['stan'].replace('很好', 'Bardzo dobry') # błąd wpisu -> kategoria

print("Braki po ujawnieniu ukrytych sentineli:")
print(df.isna().sum())
print("\nProcent braków w kolumnie (%):")
print((df.isna().mean() * 100).round(1))
````

````python
# 4. Uzupełnianie zależne (najdokładniejsze): powierzchnia * cena_m2 = cena_total
# a) brak powierzchni, ale znamy cena_m2 i cena_total -> odtwórz powierzchnię
m = df['powierzchnia'].isna() & df['cena_m2'].notna() & df['cena_total'].notna()
df.loc[m, 'powierzchnia'] = (df.loc[m, 'cena_total'] / df.loc[m, 'cena_m2']).round()
print(f"Odtworzono powierzchnię z ceny dla {m.sum()} wierszy")

# b) brak cena_m2, ale znamy powierzchnię i cena_total -> odtwórz cena_m2
m = df['cena_m2'].isna() & df['powierzchnia'].notna() & df['cena_total'].notna()
df.loc[m, 'cena_m2'] = (df.loc[m, 'cena_total'] / df.loc[m, 'powierzchnia']).round()
print(f"Odtworzono cena_m2 z powierzchni dla {m.sum()} wierszy")
````

````python
# 5. Uzupełnienie pozostałych braków wg typu zmiennej
# numeryczne dyskretne -> mediana, rzut na int (rok, piętro to liczby całkowite)
for col in ['rok_budowy', 'pietro']:
    df[col] = df[col].fillna(df[col].median()).astype(int)

# numeryczna ciągła, gdyby coś zostało -> mediana (odporna na skośność cen)
df['powierzchnia'] = df['powierzchnia'].fillna(df['powierzchnia'].median())
df['cena_m2']      = df['cena_m2'].fillna(df['cena_m2'].median())

# kategorialne binarne -> dominanta (najczęstsza wartość)
for col in ['parking', 'balkon']:
    df[col] = df[col].fillna(df[col].mode()[0])

print("Braki po pełnym uzupełnieniu:")
print(df.isna().sum())
print("\nŁączna liczba braków:", int(df.isna().sum().sum()))
````

````python
# 6. Konsekwencje metod: porównanie z prostym usunięciem wierszy (dropna)
df_raw = pd.read_csv('nieruchomosci.csv', na_values=braki_tekstowe)
print("Gdybyśmy użyli dropna() na surowych danych:")
print(f"  wierszy przed: {len(df_raw)}  ->  po dropna: {len(df_raw.dropna())}")
print(f"  STRATA: {len(df_raw) - len(df_raw.dropna())} wierszy "
      f"({(1 - len(df_raw.dropna())/len(df_raw))*100:.0f}% danych!)")

print("\nWpływ na EDA — średnia powierzchnia:")
print(f"  usunięcie (dropna):  {df_raw['powierzchnia'].dropna().mean():.1f} m² "
      f"(ale 0 i sentinele zawyżają/zaniżają!)")
print(f"  po imputacji:        {df['powierzchnia'].mean():.1f} m²")
print("\nDataFrame gotowy do analizy — typy kolumn:")
print(df.dtypes)
````

## 3. Wnioski — konsekwencje metod i wpływ na EDA

- **Usuwanie wierszy (`dropna`)** — najprostsze, ale przy tak rozproszonych brakach traci ~połowę
  danych. Nie do przyjęcia przy 44 rekordach. Dodatkowo *nie usuwa* ukrytych braków (0, -999),
  które i tak zepsułyby statystyki.
- **Imputacja wskaźnikiem centralnym** (średnia/mediana/dominanta) — zachowuje wszystkie wiersze,
  ale **zmniejsza wariancję** i może zaburzyć korelacje. Dla cen (rozkład skośny) mediana jest
  bezpieczniejsza niż średnia.
- **Imputacja zależna** (`cena_total / cena_m2`) — tu najlepsza: odtwarza wartość *dokładnie*
  z zależności biznesowej, nie wprowadzając sztucznego „spłaszczenia" rozkładu.
- **Wpływ na EDA:** kluczowe było wykrycie braków *ukrytych* (`0`, `-999`, `-1`, `很好`). Gdybyśmy
  je zostawili, średnia powierzchni, ceny i histogramy byłyby całkowicie zniekształcone
  (np. `-999` w cenie zaniża średnią o tysiące). Świadoma obsługa braków = wiarygodna analiza.
